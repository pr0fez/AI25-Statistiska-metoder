```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.column_stack([np.random.gamma(10, 7.3, size=1000),
                                   np.random.normal(0, 1, size=1000),
                                   np.random.binomial(20, 0.8, size=1000),
                                   np.random.geometric(0.3, size=1000)]),
                  columns=["A", "B", "C", "D"])
df.hist(cumulative=True)
```

```python
df.boxplot()
```

```python
import scipy . stats as stats
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 2, figsize=(10,10))
stats.probplot(adv["TV"].to_numpy(), plot=ax[0][0])
stats.probplot(df["D"].to_numpy(), plot=ax[0][1])
stats.probplot(df["B"].to_numpy(), plot=ax[1][0])
stats.probplot(adv["newspaper"].to_numpy(), plot=ax[1][1])
ax[0][0].set_title("A")
ax[0][1].set_title("B")
ax[1][0].set_title("C")
ax[1][1].set_title("D")
plt.show()
```
